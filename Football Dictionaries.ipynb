```python
# 3- grouped by country, and per each country, grouped by position :
```

```python
squads_data_grouped_By_country_position={}


for i in range(len(squads_dataD)):
    player = squads_dataD[i]
    pos = player['position']
    country = player['country']
    
    if country not in squads_data_grouped_By_country_position:
        squads_data_grouped_By_country_position[country]={}
        
    if pos not in squads_data_grouped_By_country_position[country]:
        squads_data_grouped_By_country_position[country][pos]=[]
        
    squads_data_grouped_By_country_position[country][pos].append(player)
    
```
